**Import packages**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib 
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import VarianceThreshold
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,roc_curve
```

**Read the Dataset**

```python
df=pd.read_csv(r'data_regression - data_regression.csv')
df
```

**Cheak null and empty values**

```python
df.isnull().sum()
```

**Drop values**

```python
df.dropna(inplace=True)
df.reset_index(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# change to int 
df['churn']=df['churn'].astype(int)
df
```

```python
# drop coloumns 
df.columns
```

```python
cols=['year','customer_id','phone_no']
df.drop(cols,inplace=True,axis=1)
df
```

**Label Encoder**

```python
le=LabelEncoder()
for i in df.select_dtypes(include='object'):
    df[i]=le.fit_transform(df[i])
df
```

**Divide the data into X and y**

```python
X=df.drop('churn',axis=1)
y=df['churn']
```

```python
X
```

```python
y
```

**Feature Selection**

- VarianceThreshold

```python
vt=VarianceThreshold(threshold=0)
vt.fit(X)
vt.get_support()
```

```python
vt.variances_
```

```python
cols=vt.get_feature_names_out()
X[cols]
```

**Split the data into train and test**

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**Implement the model**

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

**Make predictions**

```python
y_pred=dt.predict(X_test)
y_pred
```

**New Dataframe**

```python
df1=pd.DataFrame(X_test)
df1['y']=y
df1['y_prediction']=y_pred
df1
```

**model evalution**

```python
acc_dt=round(accuracy_score(y_test,y_pred),2)
pr_dt=round(precision_score(y_test,y_pred),2)
rc_dt=round(recall_score(y_test,y_pred),2)
f1_dt=round(f1_score(y_test,y_pred),2)
```

```python
print('accuracy score',acc_dt)
print('precision score',pr_dt)
print('recall score',rc_dt)
print('f1 score',f1_dt)
```

**Save the model**

```python
joblib.dump(dt,open('Decision_tree_model.pkl','wb'))
```
